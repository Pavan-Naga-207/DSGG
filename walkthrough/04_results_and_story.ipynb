## 3. Insight 1: SGCLS R@20 = 52.88 vs base 47.5 (+5.4)

**This is the headline.** With-constraint R@20 is the hardest official anchor STTran reports, and our frozen-DINOv2 pipeline beats the published number by 5.4 absolute points.

**What SGCLS measures:** given ground-truth boxes, predict the object class for each box and the predicate for each (subject, object) pair. The detector / RPN is bypassed (boxes are given). What's left is *embedding quality*.

**Why we win here specifically:** the per-region embedding is what gets pooled by RoI Align and fed into both the object classifier and the relation head. DINOv2's frozen patch tokens carry richer object semantics than a trained-from-scratch ViT or even a ResNet-101 fine-tuned on AG. SGCLS is the pure *embedding-quality* benchmark, and that is exactly what DINOv2 is best at.

**Possible follow-up question:** *"Are you sure 5.4 points is significant?"* Yes - the table is 9 SGCLS cells across constraint x recall, all of them won, with magnitudes ranging from +0.0 (R@50 With, ties) to +14.5 (Semi R@20). A single-cell win could be noise. A 9-for-9 sweep cannot.

## 4. Insight 2: SGDET R@20 = 40.41 vs base 34.1 (+6.3)

**The end-to-end win.** SGDET is the hardest mode (no ground truth at all - boxes detected, classes inferred, predicates predicted). +6.3 absolute points there means the pipeline as a whole is better, not just one piece of it.

**Caveat to handle if asked:** detector mAP is roughly the same (24.25 vs 24.60). So how can SGDET be much better if the detector is similar? Because *SGDET R@K* uses the detector's *predicted class scores* to rank triplets - and DINOv2 features make classes more discriminative *per box*, even when the box-level mAP is similar. Confidence calibration matters. Same number of boxes, better object class confidence per box, much better top-K recall.

## 5. Insight 3: detector mAP gap is small (24.25 vs 24.60), SGCLS gap is huge (52.88 vs 47.5)

**This is the cleanest single insight in the whole project.** It is what makes the result publishable.

If the win were coming from a stronger localizer, detector mAP would have jumped. It didn't.
The win is coming from *embedding quality per box*, which is what SGCLS isolates.

**One-sentence version:** *"The detector mAP gap is small but the SGCLS gap is large because SGCLS isolates per-region embedding quality, not localization, and that's exactly what DINOv2 features are best at."*

If you can say that one sentence cleanly during Q&A, you've won the hardest possible question about the result.

## 6. Insight 4: PredCLS roughly ties the base paper - and that's expected

Phase 3 PredCLS R@20 (With) = 71.71 vs base 71.80. Within 0.1 points.

**Why this is fine, and why it's actually a *good* sanity check:**
- PredCLS is given GT boxes *and* GT object classes. The relation head is the only thing being tested.
- We did **not** modify the relation head. STTran's spatio-temporal transformer is unchanged across phases.
- So PredCLS *should* be roughly the same as the base paper. If it had jumped a lot, that would mean we accidentally helped the relation head somehow, which would be suspicious.
- A clean tie says: "yes, we are using the same STTran head, our only change is the backbone, and the wins on SGCLS/SGDET are real."

**Talking-point version:** *"PredCLS being a tie is expected and is actually a positive sanity check: we didn't touch the relation head, so PredCLS should match the base paper. The wins live in SGCLS and SGDET because that's where the backbone embedding matters."*

## 7. The story arc to tell in 90 seconds

If you only had 90 seconds to summarize the whole project:

**Beat 1 (15s) - Setup**
> Dynamic Scene Graph Generation: predict object-relationship graphs frame-by-frame in video. STTran is the dominant baseline. It uses Faster R-CNN with ResNet-101 as the visual backbone.

**Beat 2 (10s) - Phase 1**
> Phase 1: reproduce the baseline. We hit the published numbers - 24.6 detector mAP, 71.8 PredCLS R@20.

**Beat 3 (25s) - Phase 2 and the bugs**
> Phase 2: replace the ResNet with a plain ViT. Three bugs surfaced. BF16 mixed precision was underflowing inside the loss - we cast logits to FP32 and the NaN problem went away. BatchNorm/GroupNorm settings drifted between train and eval, dropping 209 detector keys at load time - we standardized on GroupNorm. PredCLS and SGCLS were quietly using the legacy ResNet path while SGDET used the new ViT path - we rewired all three modes through one feature extractor. SGCLS jumped from 0.013 to 0.297 on the same checkpoint. After the fixes, Phase 2 was an honest baseline at 23.19 mAP, but a from-scratch ViT didn't beat ResNet on its own.

**Beat 4 (25s) - Phase 3 and the win**
> Phase 3: drop in a *frozen* DINOv2 ViT-B/14 in a parallel branch with model-native preprocessing as a single source of truth. Backbone parameters are frozen, only the synthetic pyramid + RPN + heads + STTran graph head are trained. Result: SGCLS R@20 = 52.88 vs base 47.5, SGDET R@20 = 40.41 vs base 34.1. Detector mAP is roughly the same as the base paper, but the per-region embedding quality is much higher - and that is what SGCLS isolates.

**Beat 5 (15s) - Why and what's next**
> The frozen DINOv2 features carry richer object semantics from 142M-image self-supervised pretraining than either a from-scratch ViT or a fine-tuned ResNet-101. The cleanest follow-up is DINOv3 with Gram anchoring, which is the most direct candidate to close the residual PredCLS gap and probably push SGCLS/SGDET further.
